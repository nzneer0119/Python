```python
import sys
import os

import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
from selenium import webdriver
import time
from tqdm import tqdm_notebook

query_txt = input ('검색할 키워드는 무엇입니까? :')
b = input ('수집 할 BLOG의 최소 수는? :')
blog = int(b)
f_dir = input("    -.파일을 저장할 폴더명만 쓰세요(예:c:\\temp\\):")


path = "D:/00_Limhs/py_data/drv/chromedriver.exe"  # Chrome Driver path..

driver = webdriver.Chrome(path)

driver.get('http://www.naver.com')
time.sleep(2)

now = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)

f_name=f_dir+s+'-'+query_txt+'.xls'

element = driver.find_element_by_id("query")
element.send_keys(query_txt)
element.submit()
time.sleep(1)

driver.find_element_by_link_text("VIEW").click()
time.sleep(1)

driver.find_element_by_link_text("블로그").click()
time.sleep(1)

driver.find_element_by_link_text("옵션").click()
time.sleep(1)


# item_li = driver.find_elements_by_css_selector('.option .txt')

# for i in range(0, len(item_li)):
#     print(item_li[i].text)
  

# # 3개월 위치
# print(item_li[10].text)

# item_li[10].click()

# 스크롤을 밑으로 내려주는 함수
def scroll_down(driver):
    driver.execute_script("window.scrollTo(0, 19431049)")
    time.sleep(1)
    
# n : 스크롤할 횟수 설정
n = int(blog / 30) - 1     # 1회 스크롤당 30개 blog 생성
i = 0
while i < n: # 이 조건이 만족되는 동안 반복 실행
    scroll_down(driver) # 스크롤 다운
    i = i + 1
    
# 블로그 글 url 수집
url_list = []
title_list = []

# url 크롤링 시작
articles = ".api_txt_lines.total_tit"
article_raw = driver.find_elements_by_css_selector(articles)
article_raw[0]

article_raw[0].get_attribute('href')

# 크롤링한 url 정제 시작
for article in article_raw:
    url = article.get_attribute('href')
    url_list.append(url)
time.sleep(1)

# 제목 크롤링 시작
for article in article_raw:
    title = article.text
    title_list.append(title)
    
    print(title)
    
print("")
print('url 갯수 : ', len(url_list))
print('title 갯수 : ', len(title_list))

n_blog = pd.DataFrame({'url' : url_list, 'title' : title_list})

n_blog.to_excel(f_name ,index=True)

print('저장이 완료 되었습니다.')
```
